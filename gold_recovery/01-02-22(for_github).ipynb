Получение золота из руды состоит из процессов флотации, первого и второго этапов очистки.<br>
Также в процессе образуются отвальные хвосты (остатки продукта с низкой концентрацией ценных металлов)<br>
Технологический процесс<br>


Rougher feed — исходное сырье<br>
Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant<br>
Xanthate — ксантогенат (промотер, или активатор флотации);<br>
Sulphate — сульфат (на данном производстве сульфид натрия);<br>
Depressant — депрессант (силикат натрия).<br>
Rougher process (англ. «грубый процесс») — флотация<br>
Rougher tails — отвальные хвосты<br>
Float banks — флотационная установка<br>
Cleaner process — очистка<br>
Rougher Au — черновой концентрат золота<br>
Final Au — финальный концентрат золота<br>
Параметры этапов<br>
air amount — объём воздуха<br>
fluid levels — уровень жидкости<br>
feed size — размер гранул сырья<br>
feed rate — скорость подачи<br>

### Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, черновом и финальном концентратах.

```python
data['sum_concentrate_rougher_input']=data['rougher.input.feed_ag']+data['rougher.input.feed_pb']+data['rougher.input.feed_au']
data['sum_concentrate_rougher_output']=data['rougher.output.concentrate_ag']+data['rougher.output.concentrate_pb']+data['rougher.output.concentrate_au']
data['sum_concentrate_primary_cleaner_output']=data['primary_cleaner.output.concentrate_ag']+data['primary_cleaner.output.concentrate_pb']+data['primary_cleaner.output.concentrate_au']
data['sum_concentrate_final_output']=data['final.output.concentrate_ag']+data['final.output.concentrate_pb']++data['final.output.concentrate_au']
```

```python
sns.displot(data[['sum_concentrate_rougher_input', 'sum_concentrate_rougher_output', 'sum_concentrate_primary_cleaner_output', 'sum_concentrate_final_output']], legend=True, kde=True, height=8, aspect=1.5)  
plt.title('Суммарная концентрация всех веществ на разных стадиях', fontsize=15)
plt.show()
```

```python
data[['sum_concentrate_rougher_input','sum_concentrate_rougher_output','sum_concentrate_primary_cleaner_output','sum_concentrate_final_output']].describe()
```

<b><i>Суммарная концентрация всех металлов увеличивается c каждым этапом.<br>Но есть много нулей в колонках 'sum_concentrate_rougher_output', 'sum_concentrate_primary_cleaner_output','sum_concentrate_final_output'.  </b></i>

Посмотрим на таблицу где суммарные концентрации равны нулю.

```python
data.loc[(data['sum_concentrate_rougher_output']==0)|(data['sum_concentrate_rougher_input']==0)|(data['sum_concentrate_primary_cleaner_output']==0)|(data['sum_concentrate_final_output']==0),['sum_concentrate_rougher_input','sum_concentrate_rougher_output','sum_concentrate_primary_cleaner_output','sum_concentrate_final_output']].head(10)
```

```python
data[['date','sum_concentrate_rougher_input','sum_concentrate_rougher_output','sum_concentrate_primary_cleaner_output','sum_concentrate_final_output']].head(10)
```

Как видим такие строки встречаются лиш в одном из столбцов, т.е. это ошибка.Так как эти колонки складываются из суммы (золото+серебро+свинец), мы не сможем восстановить их по отдельности для тестовой и тренировочной выборки. Но для корректного построения графика правильней будет восстановить это значение, умножив на корректирующий коэффициент (так как отношение значения к соседним колонкам колеблется в определенных узких пределах).


```python
k1=data['sum_concentrate_rougher_output'].mean()/data['sum_concentrate_rougher_input'].mean()
k2=data['sum_concentrate_primary_cleaner_output'].mean()/data['sum_concentrate_rougher_output'].mean()
k3=data['sum_concentrate_final_output'].mean()/data['sum_concentrate_primary_cleaner_output'].mean()
print(k1,';',k2,';',k3)


#Заменим нулевые значения
data.loc[(data['sum_concentrate_rougher_output']==0),'sum_concentrate_rougher_output']=data.loc[(data['sum_concentrate_rougher_output']==0),'sum_concentrate_rougher_input']*k1
data.loc[(data['sum_concentrate_primary_cleaner_output']==0),'sum_concentrate_primary_cleaner_output']=data.loc[(data['sum_concentrate_primary_cleaner_output']==0),'sum_concentrate_rougher_output']*k2
data.loc[(data['sum_concentrate_final_output']==0),'sum_concentrate_final_output']=data.loc[(data['sum_concentrate_final_output']==0),'sum_concentrate_primary_cleaner_output']*k3
```

### Напишем функцию для вычисления итоговой sMAPE

```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))

#Проверим работу функции
A=pd.Series([2,3,4,5,6,7,8,9])
F=pd.Series([1,3,5,4,6,7,10,7])
print(smape(A, F))
```

### Обучим разные модели и оценим их качество кросс-валидацией

#### Подготовка данных

Напишем функцию для вычисления итоговой sMAPE:

Приведем data_train к data_test по столбцам:

```python
#Разобьем выборку на featuress и target
#Уберем колонку "date", так как с ней модель не обучается, пусть даже 

features_train=data_train.drop(['final.output.recovery','rougher.output.recovery','date'],axis=1)
target_train = data_train[['final.output.recovery','rougher.output.recovery']]
```

Напишем фукнцию для smape, которая будет работать с двумя столбцами predictions

```python
#
def smape(targets, prediction):
    
    prediction = pd.DataFrame(prediction, columns=['final.output.recovery','rougher.output.recovery'])
    
    targets['rougher.output.recovery_predicted'] = prediction['rougher.output.recovery']
    targets['final.output.recovery_predicted'] = prediction['final.output.recovery']
    
    n = targets['rougher.output.recovery'].count()
    
    smape_rougher = (targets.apply(
        lambda x: ((np.abs(x['rougher.output.recovery'] - x['rougher.output.recovery_predicted'])) / 
                   ((np.abs(x['rougher.output.recovery']) + np.abs(x['rougher.output.recovery_predicted'])) / 2)), axis=1)).sum() / n * 100
    
    smape_final = (targets.apply(
        lambda x: ((np.abs(x['final.output.recovery'] - x['final.output.recovery_predicted'])) / 
                   ((np.abs(x['final.output.recovery']) + np.abs(x['final.output.recovery_predicted'])) / 2)), axis=1)).sum() / n * 100
   
    
    smape = 0.25 * smape_rougher + 0.75 * smape_final
    return smape
```

Предустановленные в sllearn метрики качества нам не подходят, поэтому добавим свою

```python
scorer = make_scorer(smape, greater_is_better=False) 
```

#### Линейная регрессия

```python
%%time
model = LinearRegression()
parameters = {'normalize':[False]}
rnd_lin_reg=RandomizedSearchCV(model,parameters, cv=5, random_state=12345, scoring=scorer, n_jobs=-1)
rnd_lin_reg.fit(features_train, target_train)
rnd_lin_reg.cv_results_

print('sMAPE на обучении: ',rnd_lin_reg.best_score_*-1)
```

```python
%%time
model_lin_regr = LinearRegression()
model_lin_regr.fit(features_train, target_train)
lin_reg_prd=model_lin_regr.predict(features_test)
smape_final_test=smape(target_test,lin_reg_prd)
print('Итоговоеb sMAPE для тестовой выборки: ',smape_final_test)
```

#### Случайный лес

```python
%%time
parameters = {'max_depth':range(1,80,1), 'n_estimators':range(40,60,1)}
rgr_rand_frst = RandomizedSearchCV(RandomForestRegressor (random_state = 12345), parameters, cv=5, random_state=12345, scoring=scorer, n_jobs=-1)
rgr_rand_frst.fit(features_train, target_train)
print('Лучшие параметры: ',rgr_rand_frst.best_params_)
print('sMAPE на обучении случайны лес: ',rgr_rand_frst.best_score_*-1)
```

Лучшая модель - случайный лес.

#### Проверим модель на тестовой выборке
